```python
import matplotlib.pyplot as pit
import numpy as np
from sklearn import datasets, linear_model, model_selection

# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shape of the data and the first row
print(X.shape)
print(X[0])
```

```python
# Extract the column at index 2
X = X[:, 2]
print(X.shape)

#Reshape to a 2D array
X = X.reshape((-1, 1))
print(X.shape)
```

```python
# Split the data into training and testing data
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Create a linear regression mode and train it with our data
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict using our test data
y_pred = model.predict(X_test)
```

```python
# Create a scatter plot
pit.scatter(X_test, y_test, color='black')

# Plot the predictions
pit.plot(X_test, y_pred, color='blue', linewidth=3)

# Add labels and a title
pit.xlabel('Scale BMIs')
pit.ylabel('Disease Progression')
pit.title('A Graph Plot Showing Diabetes Profession Against BMI')

# Draw the plot
pit.show()
```

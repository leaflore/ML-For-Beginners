```python
import pandas as pd

# Read the US pumpkins CSV file
pumpkins = pd.read_csv('../data/US-pumpkins.csv')

# Filter out any rows that don't use price per bushel
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
pumpkins.head()
```

```python
# This syntax Counts how many cells in each column are empty
pumpkins.isnull().sum()
```

```python
# Define the columns we want to keep
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date']

# Drop all other columns
# The drop function from the pandas library is what is being used
# the 'axis' variable at the end of the method collapses the columns of each row that we dont want
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1)
```

```python
# Calculate the average price from the high and low price columns
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2

# Get the month from the date column
month = pd.DatetimeIndex(pumpkins['Date']).month

#Create a new data frame from with these new columns
new_pumpkins = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'],'High Price': pumpkins['High Price']})
```

```python
# Convert the price of all cells prices by 1 1/9 bushels by dividing by 1 1/9
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)

# Convert the price of all cells prices by 1/2 bushels by dividing by 1/2
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)
```

```python
# Print the first few rows for the data frame
# Default amount of rows to be printed will be five(5) rows.
new_pumpkins.head()
```

```python
# Print the last few rows from the data frame
# Default amount of rows to be printed are five(5) rows.
new_pumpkins.tail()
```

```python
import matplotlib.pyplot as plt

# Get the values we want to plot
price = new_pumpkins.Price #Price is the average price between the High and Low Price
month = new_pumpkins.Month

# Create and show a scatter plot of price vs month
plt.scatter(price, month)
plt.show()
```

```python
# Group the pumpkins into a bar chart by month and price
# Used the data frame api 
new_pumpkins.groupby(['Month'])['Price'].mean().plot(kind='bar')

# Add a label to the bar chart
plt.ylabel("Pumpkin Price")
```
